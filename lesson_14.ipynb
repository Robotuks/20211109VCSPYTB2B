# Seaborn

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.histplot(data=tips, x="size")
```

```python
sns.histplot(data=tips, x="size", hue='smoker', multiple="dodge")
```

```python
sns.jointplot(data=tips, x='size', y='total_bill', kind="reg")
```

```python
sns.lmplot(data=tips, x='size', y='total_bill', hue='sex', col="day")
```

```python
sns.countplot(data=tips, x='day', hue='sex')
```

```python
flights = sns.load_dataset("flights")
flights
```

```python
flights = flights.pivot("month", "year", "passengers")
```

```python

ax = sns.heatmap(flights, cmap=sns.color_palette("ch:s=.25,rot=-.25", as_cmap=True) )
ax.set_title('Heatmap of passengers')
```
